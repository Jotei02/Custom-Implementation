```python
import numpy as np 
import matplotlib.pyplot as plt
import src.util as util
from src.p01b_logreg import LogisticRegression
```

```python
x_train1,y_train1 =util.load_dataset('data/ds1_train.csv',add_intercept=True)
x_valid1,y_valid1 =util.load_dataset('data/ds1_valid.csv',add_intercept=True)

log_reg1 = LogisticRegression()
log_reg1.fit(x_train1,y_train1)
util.plot(x_train1,y_train1,log_reg1.get_theta())
```
